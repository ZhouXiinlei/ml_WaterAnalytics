```python
import numpy as np
import pandas as pd
from datetime import datetime, date 


#可视化
import matplotlib.pyplot as plt
import seaborn as sns

#取消警告
import warnings 
warnings.filterwarnings('ignore') 
```

```python
df = pd.read_csv('dataset\Aquifer\Aquifer_Luco.csv')
df.head()
```

```python
#将 "Date" 列设置为索引，并将其转换为 DateTime 类型
df['Date'] = pd.to_datetime(df['Date'], format = '%d/%m/%Y')
df.head()
```

```python
#Depth_to_Groundwater_Podere_Casetta数据在2008-02-21才开始出现，且前面的数据有大量缺失
df1 = df.drop(df.index[0:2973])
df1.head()
```

```python

f,ax = plt.subplots(nrows=21,ncols=1,figsize=(20,40))


for i,column in enumerate(df1.drop('Date',axis=1).columns):
    sns.lineplot(x=df1['Date'],y=df1[column].fillna(method='ffill'),ax=ax[i])
    ax[i].set_title('Feature:{}'.format(column),fontsize=14)
    ax[i].set_ylabel(ylabel=column,fontsize=14)
    ax[i].set_xlabel(xlabel='Date',fontsize=14)

    ax[i].set_xlim([date(2008, 2, 21), date(2020, 6, 30)])        

plt.tight_layout()
```

```python
#后面还有缺失数据
df1 = df1.drop(df.index[6951:])
df1.head()
```

```python

f,ax = plt.subplots(nrows=21,ncols=1,figsize=(20,40))


for i,column in enumerate(df1.drop('Date',axis=1).columns):
    sns.lineplot(x=df1['Date'],y=df1[column].fillna(method='ffill'),ax=ax[i])
    ax[i].set_title('Feature:{}'.format(column),fontsize=14)
    ax[i].set_ylabel(ylabel=column,fontsize=14)
    ax[i].set_xlabel(xlabel='Date',fontsize=14)

    ax[i].set_xlim([date(2008, 2, 21), date(2019, 1, 13)])        

plt.tight_layout()
```

```python
df1.isnull().sum()
```

```python
#可以看出，仍然有大量的缺失值，把仍然有大量缺失值的列去除
df1 = df1.drop(['Rainfall_Siena_Poggio_al_Vento','Rainfall_Mensano','Rainfall_Monticiano_la_Pineta','Rainfall_Ponte_Orgia','Rainfall_Scorgiano','Rainfall_Pentolina',
                'Rainfall_Monteroni_Arbia_Biena','Depth_to_Groundwater_Pozzo_1','Depth_to_Groundwater_Pozzo_3','Depth_to_Groundwater_Pozzo_4',
                'Volume_Pozzo_1','Volume_Pozzo_3','Volume_Pozzo_4','Temperature_Mensano','Temperature_Siena_Poggio_al_Vento'],axis=1)
df1['Temperature_Pentolina'] = df1['Temperature_Pentolina'].replace(0, np.nan)
df1['Temperature_Monteroni_Arbia_Biena'] = df1['Temperature_Monteroni_Arbia_Biena'].replace(0, np.nan)
df1.isnull().sum()
```

```python
f,ax = plt.subplots(nrows=6,ncols=1,figsize=(20,15))


for i,column in enumerate(df1.drop('Date',axis=1).columns):
    sns.lineplot(x=df1['Date'],y=df1[column].fillna(method='ffill'),ax=ax[i])
    ax[i].set_title('Feature:{}'.format(column),fontsize=14)
    ax[i].set_ylabel(ylabel=column,fontsize=14)
    ax[i].set_xlabel(xlabel='Date',fontsize=14)

    ax[i].set_xlim([date(2008, 2, 21), date(2019, 1, 13)])        

plt.tight_layout()
```

```python
f, ax = plt.subplots(nrows=1, ncols=1, figsize=(16,5))

sns.heatmap(df1.T.isna(), cmap='Blues')
ax.set_title('Missing Values', fontsize=16)

for tick in ax.yaxis.get_major_ticks():
    tick.label.set_fontsize(14)
plt.show()
```

```python
#使用四种不同的方式处理缺失值
f, ax = plt.subplots(nrows=4, ncols=1, figsize=(15, 12))
#插入0
ax[0].plot(df1['Date'],df1['Rainfall_Simignano'].fillna(0),color='darkorange', label = 'modified')
ax[0].plot(df1['Date'],df1['Rainfall_Simignano'].fillna(np.inf),color='dodgerblue', label = 'original')
ax[0].set_title('Fill NaN with 0', fontsize=14)
ax[0].set_ylabel(ylabel='Rainfall_Simignano', fontsize=14)
#插入均值
mean_drainage = df1['Rainfall_Simignano'].mean()
ax[1].plot(df1['Date'],df1['Rainfall_Simignano'].fillna(mean_drainage),color='darkorange', label = 'modified')
ax[1].plot(df1['Date'],df1['Rainfall_Simignano'].fillna(np.inf),color='dodgerblue', label = 'original')
ax[1].set_title(f'Fill NaN with Mean Value ({mean_drainage:.0f})', fontsize=14)
ax[1].set_ylabel(ylabel='Rainfall_Simignano', fontsize=14)
#向前填充
ax[2].plot(df1['Date'],df1['Rainfall_Simignano'].ffill(),color='darkorange', label = 'modified')
ax[2].plot(df1['Date'],df1['Rainfall_Simignano'].fillna(np.inf),color='dodgerblue', label = 'original')
ax[2].set_title(f'FFill', fontsize=14)
ax[2].set_ylabel(ylabel='Rainfall_Simignano', fontsize=14)
#线性插值
ax[3].plot(df1['Date'],df1['Rainfall_Simignano'].interpolate(),color='darkorange', label = 'modified')
ax[3].plot(df1['Date'],df1['Rainfall_Simignano'].fillna(np.inf),color='dodgerblue', label = 'original')
ax[3].set_title(f'Interpolate', fontsize=14)
ax[3].set_ylabel(ylabel='Rainfall_Simignano', fontsize=14)

for i in range(4):
    ax[i].set_xlim([date(2008, 2, 21), date(2019, 1, 13)])
    
plt.tight_layout()
plt.show()
```

```python
df1['Temperature_Pentolina'] = df1['Temperature_Pentolina'].fillna(df1['Temperature_Pentolina'].mean())
df1['Temperature_Monteroni_Arbia_Biena'] = df1['Temperature_Monteroni_Arbia_Biena'].interpolate()
df1['Depth_to_Groundwater_Podere_Casetta'] = df1['Depth_to_Groundwater_Podere_Casetta'].interpolate()
df1['Rainfall_Simignano'] = df1['Rainfall_Simignano'].fillna(df1['Rainfall_Simignano'].mean())
df1['Rainfall_Montalcinello'] = df1['Rainfall_Montalcinello'].fillna(df1['Rainfall_Montalcinello'].mean())
df1['Rainfall_Sovicille'] = df1['Rainfall_Sovicille'].fillna(df1['Rainfall_Sovicille'].mean())
```

```python
f, ax = plt.subplots(nrows=1, ncols=1, figsize=(16,5))

sns.heatmap(df1.T.isna(), cmap='Blues')
ax.set_title('Missing Values', fontsize=16)

for tick in ax.yaxis.get_major_ticks():
    tick.label.set_fontsize(14)
plt.show()
```
